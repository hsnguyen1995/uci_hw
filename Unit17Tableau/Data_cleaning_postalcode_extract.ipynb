```python
# Import dependencies
import pandas as pd 
import glob
import requests
import json
from config import gkey

# Get folder path and csv files
path = r'C:\Users\HSN\Desktop\UCI\uci_hw\Unit17Tableau\Resources'
all_csv = glob.glob(path + "/*.csv")
```

```python
# Create empty list for both CSV files
csv_list = []

# Loop through files to append into list
for csv in all_csv:
    df = pd.read_csv(csv, index_col=None, header=0)
    csv_list.append(df)
    
# Concatenate both files
combined_csv = pd.concat(csv_list, axis=0, ignore_index=True)
```

```python
combined_csv.count()
```

```python
# Null count
combined_csv.isnull().sum().sum()
```

```python
# Drop all rows containing NaN
df = combined_csv.dropna()
```

```python
df.count()
```

```python
# Lat and Long isolation
startlatlong = df[['start station latitude', 'start station longitude']]
```

```python
# Get unique coordinates
startlatlong = startlatlong.groupby(['start station latitude', 'start station longitude']).size().reset_index(name='Freq')
startlatlong
```

```python
# Build google api URL
base_url = "https://maps.googleapis.com/maps/api/geocode/json?"
params={
    "key": gkey
}

# create empty list to append postal codes
pcode = []
count = 0

# iterate through each row to pass coordinates into the google api and grab postal codes from json results
for index, row in startlatlong.iterrows():
    lat = row['start station latitude']
    lon = row['start station longitude']
    
    params["latlng"] = f"{lat}, {lon}"
    
    response = requests.get(base_url, params=params).json()
    
    results = response["results"]
    
    pcode_extract = results[0]["address_components"][-1]["long_name"]
    
    pcode.append(pcode_extract)
    
    count += 1
    
    print(f"{count}: {pcode_extract}")
```

```python
# add postalcode list to dataframe
startlatlong['PostalCode'] = pcode
startlatlong = startlatlong[['start station latitude', 'start station longitude', 'PostalCode']]
startlatlong
```

```python
# Left merge the original and startlatlong dataframes to assign the appropriate postal code for each coordinate
final_df = pd.merge(df, startlatlong, how='left', on=['start station latitude', 'start station longitude'])
final_df
```

```python
# found a weird outlier. This station is in Canada
outlier = final_df.loc[(final_df["PostalCode"] == "H3A 1N9"), :]
print(outlier)
```

```python
# Export dataframe as csv file
final_df.to_csv("2019citibike.csv", encoding='utf-8', header=True, index=False)
```
