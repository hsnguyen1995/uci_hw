# Import Dependencies

```python
import pandas as pd 
from sqlalchemy import create_engine
from config import user, password
```

# Extract CSVs into DataFrames

```python
beer_review_file = "Resources/beer_reviews.csv"
breweries="Resources/data_world/breweries.csv"
beer="Resources/data_world/beers-cleaned.csv"
categories='Resources/data_world/categories.csv'
styles="Resources/data_world/styles.csv"
```

```python
beer_review_df = pd.read_csv(beer_review_file)
beer_review_df.head()
```

```python
brew_df=pd.read_csv(breweries)
brew_df.head()
```

```python
beer_df=pd.read_csv(beer, encoding = "ISO-8859-1")
beer_df.head()
```

```python
cat_df=pd.read_csv(categories)
cat_df
```

```python
styles_df=pd.read_csv(styles)
styles_df.head()
```

# Transform "beer_reviews" DataFrame

```python
avg_review = beer_review_df.groupby("brewery_name")["review_overall"].mean()
review_summary = pd.DataFrame({"review_overall": avg_review})
index_review_data = review_summary.reset_index()
final_review_data = index_review_data.reset_index()
final_review_data = final_review_data.rename(columns={"index": "id"})
final_review_data = final_review_data[["id", "review_overall"]]
final_review_data.set_index("id", inplace=True)
final_review_data.head()
```

# Transform "breweries" DataFrame

```python
columns=['id', 'name', 'address1', 'city', 'state', 'country', 'phone', 'website']
brew_reduced=brew_df.loc[:, columns]
brew_reduced = brew_reduced.rename(columns={"name": "brewery_name"})
brew_reduced.set_index("id", inplace=True)
brew_reduced.head()
```

# Transform "beers-cleaned" DataFrame

```python
columns=["id", 'name', 'brewery_id', 'cat_id', 'style_id', 'abv']
beer_reduced_df=beer_df.loc[:, columns]
beer_reduced_df.set_index("id", inplace=True)
beer_reduced_df.head()
```

# Transform "categories" DataFrame 

```python
cat_df.set_index("id", inplace=True)
cat_df
```

# Transform "styles" DataFrame

```python
columns=["id", 'cat_id', 'style_name', 'ABV_Low', 'ABV_High', 'Glass Type']
styles_reduced_df=styles_df.loc[:, columns]
styles_reduced_df = styles_reduced_df.rename(columns={"ABV_Low": "abv_low", "ABV_High": "abv_high", "Glass Type": "glass_type"})
styles_reduced_df.set_index("id", inplace=True)
styles_reduced_df.head()
```

# Create database connection

```python
connection_string = "postgres:postgres@localhost:5432/brewery_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm that there are tables in database
engine.table_names()
```

# Load DataFrames into database

```python
final_review_data.to_sql(name='beer_reviews', con=engine, if_exists='append', index=True)
```

```python
brew_reduced.to_sql(name='breweries', con=engine, if_exists='append', index=True)
```

```python
beer_reduced_df.to_sql(name='beers', con=engine, if_exists='append', index=True)
```

```python
cat_df.to_sql(name='categories', con=engine, if_exists='append', index=True)
```

```python
styles_reduced_df.to_sql(name='styles', con=engine, if_exists='append', index=True)
```
