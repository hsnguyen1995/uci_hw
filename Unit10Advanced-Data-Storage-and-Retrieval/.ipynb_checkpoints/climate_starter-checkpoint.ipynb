# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
first_row_M = session.query(Measurement).first()
first_row_M.__dict__
```

```python
first_row_S = session.query(Station).first()
first_row_S.__dict__
```

```python
from sqlalchemy import inspect
inspector = inspect(engine)
columns = inspector.get_columns("measurement")
for c in columns:
    print(c["name"], c["type"])
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from the last data point in the database
lastDataPoint = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
dateYearAgo = dt.date(2017,8,23) - dt.timedelta(days=365)

print(lastDataPoint)
print(dateYearAgo)
```

```python
# Perform a query to retrieve the data and precipitation scores
prcp_data = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= dateYearAgo, Measurement.prcp != None).all()
```

```python
# Save the query results as a Pandas DataFrame 
prcp_df = pd.DataFrame(prcp_data, columns=['Date', 'Precipitation'])
prcp_df.head()
```

```python
# Sort the dataframe by date and set the index to the date column
prcp_df = prcp_df.sort_values(by='Date', ascending=True)
prcp_df.set_index('Date', inplace=True)
prcp_df.head(20)
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
prcp_df.plot(figsize=(8,6), title="Precipitation Over Time")
plt.ylabel("Precipitation")
plt.tight_layout()
plt.savefig("Images_hw/precipitation.png")
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
prcp_df.describe()
```

```python
# Design a query to show how many stations are available in this dataset?
station_count = session.query(Measurement.station).distinct().count()
station_count
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in desceactivending order.
activeStations = session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
activeStations
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
temp_summary = session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).filter(Measurement.station == 'USC00519281').order_by(func.min(Measurement.tobs)).all()
temp_summary
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
station_with_highest_tobs = session.query(Measurement.tobs).filter(Measurement.station == 'USC00519281', Measurement.date >= dateYearAgo).order_by(Measurement.tobs).all()
highest_tobs_df = pd.DataFrame(station_with_highest_tobs, columns=['tobs'])
highest_tobs_df.head()
```

```python
# Plot results
highest_tobs_df.plot.hist(station_with_highest_tobs, bins=10, alpha=0.75)
plt.ylabel("Frequency")
plt.ylim(0,80)
plt.tight_layout()
plt.savefig("Images_hw/station-histogram.png")
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
prev_start_date = dt.date(2017,1,1) - dt.timedelta(days=365)
prev_end_date = prev_start_date + dt.timedelta(days=7)
previous_year_temps = calc_temps(prev_start_date, prev_end_date)
previous_year_temps
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
trip_avg_temp = list(np.ravel(previous_year_temps))
tmax = trip_avg_temp[2]
tavg = trip_avg_temp[1]
tmin = trip_avg_temp[0]
```

```python
plt.figure(figsize=(4,10))
plt.bar(1, tavg, color='coral', alpha=0.5)
plt.title("Trip Avg Temp")
plt.ylabel("Temp (F)")
plt.errorbar(1, tavg, yerr=(tmax - tmin), color='black')
plt.ylim(0,100)
plt.xticks([])
plt.tight_layout()
plt.savefig("Images_hw/temperature.png")
```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation

rainfall_per_station = session.query(Station.station, Station.name, Station.latitude, Station.longitude, Station.elevation, func.sum(Measurement.prcp)).filter(Measurement.date >= prev_start_date, Measurement.date <= prev_end_date).filter(Measurement.station == Station.station).filter(Measurement.prcp != None).group_by(Measurement.station).order_by(func.sum(Measurement.prcp).desc()).all()
print(rainfall_per_station)
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
